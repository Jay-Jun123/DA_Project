# UDEMY COURSES DATASET

```python
import pandas as pd
```

```python
courses = pd.read_csv(r'./data/7. Udemy Courses.csv')
courses
```

## 1. What are all different subjects for which Udecmy is offering courses?

```python
courses['subject'].unique()
```

## 2. Which subject has the maximum number of courses?

```python
courses['subject'].value_counts()
```

## 3. Show all the courses which are Free of Cost.

```python
courses[courses['price']=='Free'].sample(10)
```

```python
courses.loc[courses['price']=='Free','subject'].value_counts()
```

## 4. Show all the courses which are Paid.

```python
courses[courses['is_paid']==True].sample(10)
```

## 5. Which are Top Selling Courses?

```python
courses.sort_values('num_subscribers', ascending=False).head(10)
```

## 6. Which are Least Selling Courses?

```python
courses[courses['num_subscribers']>0].head(10)
```

## 7. show all courses of Graphic Design where the price is below 100? 
- ...? 문자형이고 Free가 섞임

```python
courses[(courses['subject']=='Graphic Design') & (courses['price']<'100')]
```

```python
courses[(courses['subject']=='Graphic Design') & (courses['price']=='Free')].head(10)
```

## 8. List out all the courses that are related with 'Python'.

```python
courses[courses['course_title'].str.contains('Python')]
```

## 9. What are courses that published in year 2015?

```python
courses['published'] = pd.to_datetime(courses['published_timestamp'])
courses['year'] = courses['published'].dt.year
courses[courses['year']==2015]
```

## 10. What are the Max. Number of Subscribers for Each Level of couses?

```python
courses.groupby('level')['num_subscribers'].max()
```
